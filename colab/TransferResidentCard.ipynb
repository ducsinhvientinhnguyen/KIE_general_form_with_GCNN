```python
import json
```

```python
Labels = {}
```

```python
for k in range(10):
  f = open('/content/drive/MyDrive/test/' + str(k+1) + '.json')
  L = json.load(f)
  cells = []
  for i in range(len(L['word'])):
    poly = []
    poly.append(L['word'][i]['x1'])
    poly.append(L['word'][i]['y1'])
    poly.append(L['word'][i]['x2'])
    poly.append(L['word'][i]['y2'])
    poly.append(L['word'][i]['x3'])
    poly.append(L['word'][i]['y3'])
    poly.append(L['word'][i]['x4'])
    poly.append(L['word'][i]['y4'])
    dict = {"poly": poly, "cate_id": 0, "cate_text": "OTHER", "vietocr_text": L['word'][i]['text'], "group_id": i}
    cells.append(dict)
  temp = {"cells": cells, "h_origin" : L['height'], "w_origin" : L["width"]}
  temp2 = {str(k+1): temp}
  Labels.update(temp2)
```

```python
with open('Labels.json', 'w') as outfile:
    json.dump(Labels, outfile)
```
